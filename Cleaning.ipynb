## 2. Histogram:

A histogram is a graphical representation that illustrates the distribution of **numerical data** by organizing values into **specified intervals**, known as bins. Each bin corresponds to a range of values, and the height of the bars indicates the frequency of data points within that range. This visualization helps to identify patterns, such as skewness or modality in the data, making it easier to understand the underlying distribution, detect outliers, and summarize key characteristics of the dataset. Histograms are widely used in statistics and data analysis for their effectiveness in conveying the shape and spread of data.


```python
plt.figure(figsize=(10, 9))
plt.hist(df['price'], bins=range(int(df['price'].min()), int(df['price'].max()) + 1000000, 1000000),
         color='yellow', edgecolor='black')

plt.title('Price Distribution')
plt.xlabel('Price')
plt.ylabel('count')

plt.show()
```

The histogram categorizes price values into **bins** with a width of **1,000,000**, such as the range **from 5 million to 6 million**. We chose this bin width due to the **extensive price range**, which spans **from 1,750,000 to 13,300,000**. The distribution is **positively skewed**, meaning it has a longer right tail. This skewness suggests that prices below **10 million** are less likely to occur, as most of the data points cluster in the lower range, making high values increasingly rare.

## 3. Bar plot:

A bar plot, also referred to as a bar chart or bar graph, is a popular data visualization tool that represents **categorical data** through rectangular bars. Each bar corresponds to a specific category, with its height (or length) indicating the value or frequency associated with that category. Bar plots offer a straightforward and intuitive way to visualize data, making it easy to discern patterns, trends, and variations across different categories. They are commonly utilized to display counts, frequencies, percentages, or other metrics related to distinct categories, enhancing data comprehension.

We created a bar plot to illustrate the distribution of furnishing status across three categories: **furnished, semi-furnished, and unfurnished**. ## 6. Detect Outliers:
Detecting outliers involves identifying data points that significantly deviate from the majority of the dataset. Outliers can occur due to variability in the data, errors, or rare events, and they can distort statistical analyses or model performance.

```python
import numpy as np

def count_outliers(data):
    outliers_count_per_column = {}

    numeric_columns = data.select_dtypes(include=[np.number]).columns

    for column in numeric_columns:
        Q1 = data[column].quantile(0.25)
        Q3 = data[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        outliers_count = ((data[column] < lower_bound) | (data[column] > upper_bound)).sum()
        outliers_count_per_column[column] = outliers_count
    total_outliers = sum(outliers_count_per_column.values())

    return outliers_count_per_column, total_outliers


outliers_count_per_column, total_outliers = count_outliers(data)


print("Outliers per column:", outliers_count_per_column)
print("Total outliers:", total_outliers)
```

we calculated the number of outliers in each numeric column using the IQR method and stored the results. The total number of outliers across all columns was also computed and printed.

## 6. Boxplts:
are a graphical representation of the distribution of a dataset, showing the median, quartiles, and potential outliers. The box represents the interquartile range (IQR), where 50% of the data lies, with a line inside the box marking the median. Whiskers extend to show the range of the data, and any points outside the whiskers are considered outliers.

```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_boxplots(data):

    numeric_columns = data.select_dtypes(include=['float64', 'int64']).columns

    for column in numeric_columns:
        plt.figure(figsize=(8, 4))
        sns.boxplot(data[column])
        plt.title(f"Boxplot of {column}")
        plt.xlabel(column)
        plt.show()

plot_boxplots(data)
```

we created boxplots for each numeric column in the dataset using Seaborn to visualize the distribution and detect outliers. Each plot displays the range, median, and potential outliers for the respective column.

## 6. Handling Outliers:
involves identifying and addressing data points that significantly differ from the rest of the dataset, as they can skew analysis results. Common techniques include removing outliers, capping them to the nearest non-outlier values, or transforming the data to reduce their impact. Properly managing outliers is crucial for improving model accuracy and ensuring reliable statistical conclusions.

```python
import pandas as pd

outlier_threshold = 1.5

def count_outliers_and_cap(column_data, threshold=1.5):
    Q1 = column_data.quantile(0.25)
    Q3 = column_data.quantile(0.75)
    IQR = Q3 - Q1

    upper_bound = Q3 + threshold * IQR
    lower_bound = Q1 - threshold * IQR

    outliers = (column_data > upper_bound) | (column_data < lower_bound)
    outlier_count = outliers.sum()

    capped_column = column_data.clip(lower=lower_bound, upper=upper_bound)

    return capped_column, outlier_count

outlier_counts = {}
total_rows = len(data)
numeric_columns = data.select_dtypes(include=[np.number]).columns

for column in numeric_columns:
    capped_column, outliers = count_outliers_and_cap(data[column], threshold=outlier_threshold)

    data[column] = capped_column

    outlier_counts[column] = outliers
    data.to_csv('Housing_cleaned.csv', index=False)
```

 we defined a function to count and cap outliers in each numeric column of the dataset using the IQR method. The capped values replaced the original data, and the cleaned dataset was then saved to a new CSV file, Housing_cleaned.csv.
