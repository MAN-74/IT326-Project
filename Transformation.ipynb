```python
import pandas as pd
from google.colab import files
Data = files.upload()
df = pd.read_csv(next(iter(Data)))
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

## **Data Transmation:**

**Encoding:**

```python
#Encoding

# Converting the `furnishingstatus` column
df['furnishingstatus'] = encoder.fit_transform(df['furnishingstatus'])
df['furnishingstatus'] = df['furnishingstatus'].replace({0: '3', 1: '2', 2: '1'})

# Converting the 'mainroad' column
df['mainroad'] = df['mainroad'].replace({'yes': 1, 'no': 0})

# Converting the 'guestroom' column
df['guestroom'] = df['guestroom'].replace({'yes': 1, 'no': 0})

# Converting the 'basement' column
df['basement'] = df['basement'].replace({'yes': 1, 'no': 0})

# Converting the 'hotwaterheating' column
df['hotwaterheating'] = df['hotwaterheating'].replace({'yes': 1, 'no': 0})

# Converting the 'airconditioning' column
df['airconditioning'] = df['airconditioning'].replace({'yes': 1, 'no': 0})

df['prefarea'] = df['prefarea'].replace({'yes': 1, 'no': 0})
```

```python
from IPython.display import display
pd.set_option('display.max_row' ,None)
pd.set_option('display.max_columns' ,None)
display(df)

```

Encoding Categorical Data We applied encoding techniques to transform categorical into factors, making it easier for the model to interpret the data. This transformation facilitates the implementation of machine learning algorithms and simplifies statistical analysis. By converting categories into numerical representations, we enable the model to process the information efficiently.

**Aggregation:**

```python
avgpperb = df.groupby('bedrooms')['price'].mean()
print(avgpperb)
```

The Aggregation technique was used to calculate the average price of houses based on the number of bedrooms. By grouping the data according to the number of bedrooms column, I was able to obtain the average prices for each category of bedrooms, which helps in understanding market trends.

**Discretization:**

```python
breaks_price = df['price'].quantile([0, 0.5, 1])


df['price'] = pd.cut(df['price'], bins=breaks_price, labels=['Low', 'High'], include_lowest=True)

display(df)
```

The Discretization technique was employed on the class label price to simplify and enhance the interpretation of large continuous values by grouping them into specific intervals. This approach facilitates the analysis and communication of the models results. A new column was added to categorize price into two groups: Low and High. The Low category represents prices that are less than the median, while the High category includes prices that are equal to or greater than the median. This process involved using the quantile() function to identify breakpoints, followed by the cut() function to discretize the prices based on those breakpoints, with the labels designated as Low and High.

**Normalization:**

```python
#Normaliztion

from sklearn.preprocessing import MinMaxScaler

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

scaler = MinMaxScaler(feature_range=(0.1, 1))

df['area'] = scaler.fit_transform(df[['area']])

display(df)
```

The area attribute was rescaled using the Normalization technique to fall within a range of 0 to 1. This method standardizes feature values, ensuring that all values are on the same scale without distorting the differences between ranges. Normalization is essential in algorithms that are sensitive to the scale of data
